### ðŸ“˜ Key Terms You Should Know

- task: Specifies the type of problem. Common values:
  - detect â€“ Object detection (bounding boxes)
  - segment â€“ Object segmentation (pixel-level masks)
  - classify â€“ Image classification
- mode: Specifies what you want to do:
  - train â€“ Train a model from scratch or checkpoint
  - predict â€“ Run inference on image, video, webcam
  - val â€“ Validate model accuracy on test set
- model: Path to the YOLOv8 .pt file (e.g., yolov8n.pt)
- source: The input data (e.g., camera ID, image path, video file, URL)

### ðŸ“· Step 3: Run Real-Time Detection Using a Camera

Detect objects from a USB or CSI camera:

```python
yolo task=detect mode=predict model=yolov8n.pt source=0
```

---

### ðŸ–¼ Step 4: Run Detection on an Image

Use any local `.jpg` or `.png` file:


```python
yolo task=detect mode=predict model=yolov8n.pt source='path/to/image.jpg'
```

---

### ðŸŽž Step 5: Run Detection on a Video File
